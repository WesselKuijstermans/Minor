```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv('testfile_somnox_labeled.csv')

df = df[df['label'] != 'noise']

# reset the index to between 0 and the length of the dataframe
df = df.reset_index(drop=True)

scaler = StandardScaler()
df['somnox_ax'] = scaler.fit_transform(df['somnox_ax'].values.reshape(-1, 1))

# split df into train and test. 70% train, 30% test. The split is before and after an index
train = df.loc[:int(len(df)*0.7)]
train_x = train[['somnox_ax', 'time']]
train_y = train['label']
test = df.loc[int(len(df)*0.7):]
test_x = test[['somnox_ax', 'time']]
test_y = test['label']
print(train.shape, test.shape)

train.head()
```

```python
# Initialize the RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model on the training data
rf.fit(train_x, train_y)

# Predict the labels of the test set
predictions_test = rf.predict(test_x)
predictions_df = rf.predict(df[['somnox_ax', 'time']])


# Calculate the accuracy of the model
accuracy = np.mean(predictions_test == test_y)
print(f'Accuracy Test: {accuracy}')

test = test.assign(predictions=predictions_test)

# save the predictions to a csv file
test.to_csv('testfile_somnox_labeled_predictions.csv')
```
