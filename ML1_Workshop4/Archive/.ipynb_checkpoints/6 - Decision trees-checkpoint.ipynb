### Data Cleaning

Cleaning the data is not the focus of this notebook, but we need to do something to get valid data. We'll impute the data with a **backfill algorithm**. This means that we fill in the missing data with *hopefully* similar data from the next row in the table.

We want to fill missing data with data from rows that are similar. Say the `deck` is missing for some row. If we fill this with a deck we want to make sure the passenger we use to fill the missing data is similar. It doesn't make much sense to say a third class passenger would be travelling on a first class deck.

You might have seen that there are still some NaN's after filling with the backfill method. If the sorting algorithm puts all the NaN's at the end we're imputing NaN's. Since that only happens for the `age` column we'll impute the last couple with the mean age of the data set.

```python
# Sort on all columns (inplace saves us an assignment)
df.sort_values(by=list(df.columns), axis=0, inplace=True)

# Fill missing data with the value of the next row (backfill)
df.fillna(method='bfill', inplace=True)

# Fill remaining gaps with mean
df.fillna(df.mean(), inplace=True)

df.head(10)
```

The decision tree of Scikit works only on numerical dimensions. So we'll have to turn all the categories into dummy variables. Remember the Pandas function for this?

```python
df = pd.get_dummies(df)
df.columns
```

```python
df.dtypes
```

### Training and Testing

We not only have to train our decision tree model, we should also verify that it works. However, we just have one dataset available to us...

With this verification we make sure that we don't **overfit** the model. We could learn to output exactly the right answer of our data set, but we might not get any new data correct. This is very similar to using very high degree polynomials in a regression fit. We'll dive into this deeper later in the notebook. The [wikipedia](https://en.wikipedia.org/wiki/Overfitting) page on overfitting is also a good resource to learn more on overfitting.

A standard solution to this problem is dividing up the data you have available into a **training set** and a **test set**. We train the model on the training set and verify that we didn't overfit on the test set. 

Let's split our imputed data set! (Note that `train_test_split` was imported from scikit at the top.)

```python
df_train, df_test = train_test_split(df, test_size=0.5, random_state=42)
print(df_train.shape, df_test.shape)
```

For now we just split the set in half. It is a randomized split of the data, however you can *seed* the randomizer with `random_state`. This means you get the *same* random split each time you run this code. (If you omit this parameter you will get a new random split every time.)

### A Manual Decision Tree

Let's start by building a decision tree manually. I promise you won't have to do this manually each time!

Which feature do you think would be a good predictor for survival? Let's pick one based on domain knowledge: the famous story goes that men were expected to let women and children board the life rafts first. 

Luckily there is a feature for this! We can define a simple function that returns the survival prediction based on the `adult_male` column.

```python
def decide_adult_male(row):
    if row.adult_male:
        return 0
    else:
        return 1
```

Now we can use this **decision tree of one deep** to predict the survival of our test set. 

```python
pred_1deep = df_test.apply(decide_adult_male, axis=1)
pred_1deep
```

We need some definition of an error to see how far off we are. Let's take a simple measure of how many rows were misclassified. 

In the function below we loop over the true values and the predicted values and add their squared difference. Since both can only be a `0` or a `1` the squared difference is `0` if the prediction matches the true value and `1` if they don't match. (Check this for yourself if you're not sure about this!)

---
#### Exercise
Did you see we use the `zip` function in the loop?

Try to figure out what the zip function does by zipping two simple lists, for example `[1,2,3]` and `['one', 'two', 'three']`.

This is actually a common way to loop over multiple lists simultaneously in Python.

---

```python
def compute_error(truth, prediction):
    diff = 0
    
    for truth_i, prediction_i in zip(truth, prediction):
        diff += (truth_i - prediction_i)**2
        
    return diff
```

Ok, with that definition let's find out how we did!

```python
compute_error(df_test['survived'], pred_1deep)
```

Oof.. We made 87 mistakes out of the 446 samples. On the other hand, we're making some good predictions with this tiny tree. 

It would be unreasonable to expect that we can predict well based on just one column. We can make our tree **two deep** by adding another layer of questions. Just to get the concept of a decision tree right we'll do these manually too. 

Let's say we first want to decide on the `class` passengers were travelling in. Perhaps first class passengers got priority when boarding the life rafts. 

After the class decision we don't have to decide on the same column. Which column we look at next can be different depending on which class a passenger was traveling in. We'll stick to our earlier decision of seeing is a passenger is an adult male when they're traveling in second or third class. But when a passenger was first class we'll predict their fate based on whether they were on this journey alone.

Below are the two new functions we need to make the prediction. Make sure you understand how the three functions we defined so far work together!

```python
def decide_alone(row):
    if row.alone:
        return 1
    else:
        return 0

def decide_class(row):
    if row.class_First:
        return decide_alone(row)
    else:
        return decide_adult_male(row)
```

Now we can compute the prediction our tree of two deep makes.

```python
pred_2deep = df_test.apply(decide_class, axis=1)
pred_2deep
```

And again we can compute how many we got wrong.

```python
compute_error(df_test['survived'], pred_2deep)
```

That's getting better than what we had before! 

It's pretty difficult though to determine which features we need to make a decision with. Maybe we can make some algorithm **learn** to make these decisions!

### Tree Learning

Now we know what the **model** looks like (a decision tree is a machine learning model just as linear or logistic regression is an ML model) we can see how to train this model.

The essence of a decision tree is that it divides the entire data set into **bins**. Each bin contains some number of data points or in our case passengers on the Titanic. The goal is to have bins where all passengers either survive the disaster or not. If we have these bins we can predict whether someone will survive by simply looking up in which bin the passenger falls and predicting the same outcome.

Let's compare this to linear regression. The goal in linear regression is to find a line that lies as close as possible to the data points. Now in linear regression you've used **gradient descent** as a method to learn the **parameters** of the model. Gradient descent minimizes the **cost function** when training the model. So what are the parameters in a decision tree? And what algorithm do we use to learn them?

The **parameters of a decision tree are the decisions** we make in the tree. So we need to learn which decision to make first, or in other words, which decision tells us most about someone's fate. 

The **cost function of a decision tree** is a somewhat complex mathematical formula that determines how different two groups are. A common function used for this is the [Gini coefficient](https://en.wikipedia.org/wiki/Gini_coefficient). We'll not go into the mathematical details, but what you need to know is that the Gini coefficient goes from 0 to 1. 0 means that samples in a bin are the same and 1 means that all samples in the bin are different. (As a side note, economists use the Gini coefficient to determine the wealth inequality in a country).

The final thing we need is an **algorithm that minimizes the cost function**. In the case of a decision tree this is often a **greedy algorithm**. Each time we need to make a new decision in the tree it just looks at all the columns we didn't use for a decision yet and tries it out. It then looks at if the Gini coefficient of the bins becomes smaller and puts the decision in the tree that makes the Gini coefficient smallest. Although this algorithm works very differently from gradient descent, the general idea is still the same. Step by step we minimize the cost function (in this case Gini coefficient).

Let's get a feel for the Gini coefficient. First we compute the average survival rate over the entire training set. 

```python
df_train['survived'].mean()
```

Now let's look at a feature that we expect to be predictive of survival. We separate passengers based on which class they were traveling in and see what the average survival rate in each class was.

```python
df_train[df_train['class_First'] == 1]['survived'].mean()
```

```python
df_train[df_train['class_Second'] == 1]['survived'].mean()
```

```python
df_train[df_train['class_Third'] == 1]['survived'].mean()
```

That separates the passengers pretty well! People in first class tended to survive the Titanic, while second class passengers were about average. If you were traveling in third class you had a significantly lower survival chance than average. 

Binning the data on this feature (or in other words, making a decision based on this feature) would therefore *decrease* the Gini coefficient, which brings us a step closer to the minimum of the cost function.

Now let's look at a feature which we think will *not* predict survival. We can split the passengers on where they embarked the Titanic on its last voyage. Probably each of the three towns had a mix of passengers coming aboard. 

```python
df_train[df_train['embarked_C'] == 1]['survived'].mean()
```

```python
df_train[df_train['embarked_Q'] == 1]['survived'].mean()
```

```python
df_train[df_train['embarked_S'] == 1]['survived'].mean()
```

The survival rates in each of the categories is quite close to the average survival rate. This means that this feature separate the passengers not that well. The Gini coefficient with this decision would be much larger than with the `class` decision. 

The class they were traveling in is therefore the decision we want to add to the tree!

### Now a Real Tree

Ok, the real deal now. Let's import the Scikit [DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html) and see how it works on the Titanic set. 

Scikit works on two data sets: a set with features and a set with target variables. Let's split the training and test set we made earlier. To create the feature set we remove the columns with information on whether the passengers lived. As target we simply want to predict whether they survive.

```python
X_train = df_train.drop(columns=['survived', 'alive_no', 'alive_yes'])
y_train = df_train['survived']
X_test = df_test.drop(columns=['survived', 'alive_no', 'alive_yes'])
y_test = df_test['survived']
```

Training the tree is then as simple as calling the `fit` method on a new tree instance:

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

Wow, training went pretty fast on this set. Now we can see what predictions our model makes. 

We want to know the predictions on the test set, but also on the training set. This is as simple as calling `predict` on the trained tree with the feature set.

```python
pred_tree_train = dt.predict(X_train)
pred_tree_test = dt.predict(X_test)
```

Let's find out how our tree performed on the test set!

```python
compute_error(y_test, pred_tree_test)
```

Hey, pretty good! We made a lot fewer mistakes than with our manual mini tree above.  Now let's see what the tree predict for the training set.

### Using Trees to Select Features

What you've done in the exercise above is actually a kind of feature selection. You were searching for which features were most telling whether someone survived. Probably you (rightly) expected to find these features higher up in the tree. 

There is actually an algorithm provided by Scikit to filter out the features that predict the target variables the best: `SelectFromModel`. We need to give this object some model to determine which features are most predictive. Let's give it a decision tree. We can then simply make it find these features by calling `fit` again.

```python
from sklearn.feature_selection import SelectFromModel
sel = SelectFromModel(DecisionTreeClassifier())
sel.fit(X_train, y_train)
```

Now we can find the *support*: all the features that are most predictive.

```python
sel.get_support()
```

Hmm... Let's make this a bit more readable and make a list of column names out of it. If we also add the survived column we can look at the correlations between the features that are most predictive.

```python
cols = list(X_train.columns[sel.get_support()])
cols += ['survived']
cols
```

### Overfitting a Tree

Finally we'll explore something that can go wrong with decision trees: **overfitting**. The titanic data set is not large enough to see a real devastating effect of overfitting, but in general overfitting becomes a bigger problem in larger data sets.

We define a function that learns a tree with a maximum depth that we pass as parameter. It then predicts the outcomes of the train and test set.

```python
def learn_tree(depth):
    dt = DecisionTreeClassifier(max_depth=depth)
    dt.fit(X_train, y_train)
    pred_train = dt.predict(X_train)
    pred_test = dt.predict(X_test)
    err_train = compute_error(y_train, pred_train)
    err_test = compute_error(y_test, pred_test)
    return err_train, err_test
```

Let's get a feel for what happens when we change the maximum depth of the tree. We pick some arbitrary numbers, a shallow tree, and average one and a deep tree.

```python
learn_tree(5)
```

```python
learn_tree(10)
```

```python
learn_tree(20)
```

First take a look at the error in the training set. Each time we increase the maximum depth of the tree there are fewer mistakes in the classification of the training set. Yay! We keep learning when we make the tree deeper!

Hold on... Now look at the number of mistakes in the test set. It didn't decrease significantly when we made the tree deeper. How's that possible?!

This is classic overfitting. We're learning more and more of the training data set, but this doesn't **generalize** to the data set as a whole. We're just learning the peculiarities of the training set instead of the underlying structure in the data. 

We can see this more clearly if we plot the number of misclassifications versus the maximum depth of the tree. First we generate the data and then plot it.

```python
err_train = []
err_test = []
depths = range(1, 40)

for depth in depths:
    err_train_i, err_test_i = learn_tree(depth)
    err_train.append(err_train_i)
    err_test.append(err_test_i)
    
plt.plot(depths, err_train, label='train')
plt.plot(depths, err_test, label='test')
plt.legend()
plt.xlabel('maximum depth')
plt.ylabel('misclassifications')
plt.show()
```

We actually didn't need such a large tree after all! Do you see how the number of mistakes doesn't decrease any more in the test set after a maximum depth of about 5? All the effort we put into learning after this is just spent overfitting the tree.

If you have a large data set the overfitting might not be so harmless as in this example. The performance on the test data could actually get worse the more you train it! So be sure to check this on large data sets.
